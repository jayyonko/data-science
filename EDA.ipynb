```python
corelation = data.corr()
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns
               ,annot=True)
```

```python
sns.pairplot(data)
```

```python
sns.relplot(x='average_montly_hours',y='satisfaction_level',hue='left',
           data=data)
```

```python
#as we can see people with higher satisfaction level are less likely to leave the company but if their average monthly
#hours are more than 200 hours they are more likely to leave. in general people with higher working hours are very likely to leave
```

```python
sns.relplot(x='time_spend_company',y='average_montly_hours',hue='left',
           data=data)
```

```python
#as we can see employees with more than 6 years of experience in the company are not likely to leave the company
```

```python
sns.relplot(x='salary',y='average_montly_hours',hue='left',
           data=data)
```

```python
#employees with low to medium salary are very likely to leave specially if their working hours are more than 200 hours
#or less than 150 hours.employees with high salary are very less likely to leave
```

```python
sns.relplot(x='promotion_last_5years',y='average_montly_hours',hue='left',
           data=data)
```

```python
#employees who got promted in last 5 years are very less likely ot leave
```

```python
sns.relplot(x='number_project',y='average_montly_hours',hue='left',
           data=data)
```

```python
#for our basic analysis we come to conclusion that 'working hours','experience in the company',and 'satisfaction level' are 
#some very crusial aspects to predict employee retention
```
